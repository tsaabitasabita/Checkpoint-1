# **Visualisasi Data**

## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Menampilkan data kunjungan website beserta sumber data

```python
df = pd.read_csv('/content/drive/MyDrive/daily-website-visitors.csv')
```

```python
print("Sumber Data: https://www.kaggle.com/datasets/bobnau/daily-website-visitors")
df.head()
```

## **Analisis Perbandingan Rata-rata Kunjungan Pertama Kali vs Kunjungan Berulang menggunakan Bar Chart**

```python
plt.figure(figsize=(12, 6))
ax = df.groupby('First.Time.Visits')['Returning.Visits'].sum().plot(kind='bar', color='red')
plt.title('Kunjungan Pertama Kali dan Kunjungan Berulang')
plt.xlabel('Kunjungan Pertama Kali')
plt.ylabel('Kunjungan Berulang')
plt.xticks(rotation=90)
plt.xticks(range(0, len(ax.get_xticks()), 5), ax.get_xticklabels()[::5])
plt.show()
```

##Insight yang didapat dari grafik:
1. Ada lonjakan pengunjung yang besar di beberapa titik tertentu.Artinya, pada waktu-waktu tertentu, banyak pengunjung yang kembali lagi ke website. Ini bisa jadi karena ada promosi atau event khusus yang menarik perhatian mereka.

2. Kunjungan berulang cukup stabil di sebagian besar waktu menunjukkan website memiliki pengunjung setia yang terus kembali, meskipun jumlahnya tidak terlalu tinggi.

3. Ada pola yang bisa dimanfaatkan untuk meningkatkan kunjungan.
Lonjakan yang terjadi bisa dijadikan acuan untuk membuat strategi agar pengunjung kembali lagi.

##Tindakan yang bisa dilakukan:
1. Cari tahu penyebab lonjakan.
Misalnya, apakah ada promosi, event, atau konten menarik yang membuat pengunjung tertarik untuk kembali.

2. Perkuat strategi di waktu-waktu yang stabil.
Berikan penawaran khusus atau konten yang menarik agar pengunjung semakin sering kembali.

3. Terapkan strategi sukses di waktu lain.
Gunakan cara yang berhasil saat lonjakan terjadi untuk meningkatkan kunjungan di waktu lain.

## **Analisis Tren Kunjungan Unik Website per Hari menggunakan Line Chart**

```python
df['Date'] = pd.to_datetime(df['Date'])
df_visits_per_day = df.groupby('Date')['Unique.Visits'].sum()
sample_data = df_visits_per_day.iloc[:30]
plt.figure(figsize=(10, 5))
plt.plot(sample_data.index, sample_data.values, marker='o', color='purple')
plt.title('Tren Jumlah Pengunjung per Hari (Sampel Data)')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Pengunjung Unik')
plt.xticks(rotation=45, fontsize=8)
plt.yticks(fontsize=8)
plt.grid()
plt.tight_layout()
plt.show()
```

##Insight yang didapat dari grafik:

1. Terlihat adanya peningkatan jumlah pengunjung unik secara bertahap dari waktu ke waktu, meskipun terdapat beberapa titik penurunan yang signifikan.
2.  Terdapat lonjakan tajam pada beberapa waktu, yang bisa menunjukkan adanya promosi atau event khusus yang meningkatkan trafik.
3.  Setelah lonjakan, jumlah pengunjung cenderung stabil, yang menunjukkan adanya retensi pengunjung yang baik.

##Tindakan yang bisa dilakukan:
1. Identifikasi faktor yang menyebabkan lonjakan trafik, seperti promosi atau kampanye digital, dan replikasi strategi tersebut untuk meningkatkan kunjungan di masa depan.
2. Fokus pada konten yang diminati pengunjung saat lonjakan terjadi untuk mempertahankan minat pengunjung.
3. Lakukan pemantauan rutin untuk memastikan kunjungan stabil, mengantisipasi penurunan drastis dengan strategi pemasaran yang tepat.

## **Analisis Korelasi Antar Jenis Kunjungan Website menggunakan Heatmap**

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df[['First.Time.Visits', 'Returning.Visits', 'Unique.Visits']].corr(), annot=True, cmap='Spectral', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

##Insight yang didapat dari grafik dan tindakan yang bisa dilakukan:
>1. Korelasi Kuat antara First Time Visits dan Unique Visits (0.97) menunjukkan bahwa jumlah kunjungan pertama kali sangat berpengaruh pada jumlah kunjungan unik. Artinya, semakin banyak kunjungan pertama kali, semakin tinggi pula jumlah pengunjung unik.

>Tindakan:
Fokus pada strategi pemasaran untuk menarik pengunjung baru, seperti promosi, diskon, atau iklan berbayar.


>2. Korelasi Sedang antara Returning Visits dan Unique Visits (0.64) menandakan bahwa kunjungan berulang juga berdampak pada jumlah kunjungan unik, meskipun tidak sekuat kunjungan pertama kali.

>Tindakan:
Tingkatkan program loyalitas atau pemberian insentif untuk mendorong pengunjung kembali lagi ke situs.


>3. Korelasi Rendah antara First Time Visits dan Returning Visits (0.45). korelasi yang lebih rendah menunjukkan bahwa pengunjung yang baru pertama kali datang tidak selalu kembali lagi.

>Tindakan:
